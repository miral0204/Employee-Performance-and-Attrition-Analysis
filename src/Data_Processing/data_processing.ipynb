### Checking for outliers
 * Checking for outliers using IQR Method for numerical columns which may have outliers present
 * Outliers affect the model performance and may skew the predictions drastically.
 * Outliers needs to be either removed or replaced with mean or median of the data for better model performance

```python
import pandas as pd

def sum_of_outliers_iqr(df, columns):
    outlier_sums = {}
    
    for col in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        
        # Calculate IQR
        IQR = Q3 - Q1
        
        # Determine outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Find outliers and their sum
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_sum = outliers[col].unique()
        
        outlier_sums[col] = outlier_sum
    
    return outlier_sums

# List of columns to check for outliers
columns_to_check = [
    'Age', 'DistanceFromHome', 'EmpHourlyRate', 'EmpLastSalaryHikePercent',
    'TotalWorkExperienceInYears', 'TrainingTimesLastYear',
    'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole',
    'YearsSinceLastPromotion', 'YearsWithCurrManager'
]

# Calculate the sum of outliers
outlier_sums = sum_of_outliers_iqr(df, columns_to_check)

# Print the sum of outliers for each column
for col, sum_value in outlier_sums.items():
    print(f"Column: {col}, Sum of Outliers: {sum_value}")
```

### Instead of removing outliers , as we have a limited dataset, we will replace it with the median of other entries.

```python
import pandas as pd

def replace_outliers(df, columns, strategy='median'):
    df_copy = df.copy()  # Create a copy to avoid modifying the original dataframe
    
    for col in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df_copy[col].quantile(0.25)
        Q3 = df_copy[col].quantile(0.75)
        
        # Calculate IQR
        IQR = Q3 - Q1
        
        # Determine outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Replace outliers with median or mean
        if strategy == 'median':
            replacement_value = df_copy[col].median()
        elif strategy == 'mean':
            replacement_value = df_copy[col].mean()
        else:
            raise ValueError("Strategy should be either 'median' or 'mean'")
        
        # Replace outliers with the chosen value
        df_copy.loc[(df_copy[col] < lower_bound) | (df_copy[col] > upper_bound), col] = replacement_value
    
    return df_copy

# List of columns to replace outliers
columns_to_replace = [
    'Age', 'DistanceFromHome', 'EmpHourlyRate', 'EmpLastSalaryHikePercent',
    'TotalWorkExperienceInYears', 'TrainingTimesLastYear',
    'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole',
    'YearsSinceLastPromotion', 'YearsWithCurrManager'
]

# Replace outliers with median
df = replace_outliers(df, columns_to_replace, strategy='median')


```

### Data transformation : Encoding Categorical Data

* #### Many fields in our dataset are categorical in nature, i.e. either they follow a rank or hierarchy or describes the distribution class of a column. But for model training we need the data to be in numerical format, hence we will compute them using LabelEncoder & OrdinalEncoder.
* #### For columns like gender , employee department etc, we do not have a specific order or magnitude(rank) , hence the model does not need to give a weighted value to each categorical value. For such cases, we use LabelEncoding to process categorical values.
* #### For columns which describe rank and the importance of a feature with its values, like Job Satisfaction , Job level, Travel Frequency , We need to keep the order intact and hence use OrdinalEncoding to process each value of a column while keeping its importance intact

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
# These columns do not follow any specific order or rank 
nominal_columns = ['Gender', 'MaritalStatus', 'EmpDepartment', 'OverTime', 'Attrition','EducationBackground','EmpJobRole']

# Apply Label Encoding 
label_encoders = {}
for col in nominal_columns:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le
    else:
        print(f"Missing nominal column: {col}")

# Ordinal columns : these columns have a specific 
ordinal_columns = {
    'BusinessTravelFrequency': ['Non-Travel', 'Travel_Rarely', 'Travel_Frequently']
}

# Apply Ordinal Encoding to ordinal columns
for col, categories in ordinal_columns.items():
    if col in df.columns:
        encoder = OrdinalEncoder(categories=[categories])
        df[col] = encoder.fit_transform(df[[col]])
    else:
        print(f"Missing ordinal column: {col}")


```
